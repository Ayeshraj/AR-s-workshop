```python
import requests
url = "https://en.wikipedia.org/wiki/Brown_bear"
result = requests.get(url)
result.text
```

```python
import bs4
soup = bs4.BeautifulSoup(result.text, 'lxml')
```

```python
soup.select("title")[0].getText()
```

```python
soup.select("p")[0]
```
